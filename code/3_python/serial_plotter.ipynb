```python
import time

import matplotlib.pyplot as plt
import serial
from IPython.display import clear_output, display
```

```python
PORT = "/dev/cu.usbserial-14140"  # Change to '/dev/ttyUSB0' or '/dev/ttyACM0' on Linux/Mac
BAUD_RATE = 9600
TIMEOUT = 1
```

```python
def live_serial_plot():
    # Initialize lists to store data
    data_points = []
    timestamps = []
    time_counter = 0

    try:
        # Open serial port
        ser = serial.Serial(PORT, BAUD_RATE, timeout=TIMEOUT)
        time.sleep(2)  # Wait for connection to stabilize

        print(f"Connected to {PORT}. Reading data...")

        # Setup the plot
        fig, ax = plt.subplots(figsize=(10, 5))

        while True:
            line = ser.readline().decode("utf-8").strip()

            if line:
                try:
                    # Convert string to float
                    # print(line)
                    float_list = [float(x) for x in line.split()]
                    data_points.append(float_list)
                    time_counter += 1
                    timestamps.append(time_counter)

                    # Keep only the last 100 points for a "scrolling" effect
                    if len(data_points) > 100:
                        data_points.pop(0)
                        timestamps.pop(0)

                    # Update the plot
                    ax.cla()

                    setpoint_vals, actual_vals = zip(*data_points)
                    ax.plot(timestamps, setpoint_vals, marker="o", color="b")
                    ax.plot(timestamps, actual_vals, marker=".", color="r")
                    ax.set_title("Real-Time Serial Data")
                    ax.set_ylabel("Value")
                    ax.set_xlabel("Sample Number")
                    ax.grid(True)

                    # Refresh the notebook display
                    clear_output(wait=True)
                    display(fig)

                except ValueError:
                    # Skip lines that aren't valid numbers
                    continue

    except KeyboardInterrupt:
        print("Live plot stopped by user.")
    except Exception as e:
        print(f"Error: {e}")
    finally:
        if "ser" in locals() and ser.is_open:
            ser.close()
            print("Serial port closed.")


# Run the function
live_serial_plot()
```
